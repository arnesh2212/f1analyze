```python
import pandas as pd

# Load data
data = pd.read_csv('train.csv')

# Remove duplicates
data = data.drop_duplicates()

# Drop unnecessary columns
columns_to_drop = ['resultId', 'racerId', 'driverId', 'constructorId', 'number', 'url_x', 'url_y',  'date', 'time_y', 
                   'fp1_date', 'fp1_time', 'fp2_date', 'fp2_time', 'fp3_date', 'fp3_time', 'quali_date',  
                   'sprint_date', 'sprint_time', 'driverRef', 'driver_num',  'forename', 'surname', 'dob', 
                   'nationality', 'constructorRef', 'nationality_y', 'url', 'status']

data = data.drop(columns=columns_to_drop)

# Handle missing values
# Assuming numeric columns with missing values are to be filled with the mean
numeric_columns = data.select_dtypes(include=['number']).columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())

# For categorical columns, fill with the mode
categorical_columns = data.select_dtypes(include=['object']).columns
data[categorical_columns] = data[categorical_columns].fillna(data[categorical_columns].mode().iloc[0])

# Convert data types
# Ensuring numeric columns are in the correct format
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric)

data.head()
```
